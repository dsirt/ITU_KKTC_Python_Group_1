```python
#Random Sayı Üretimi
from tkinter import *
import random

liste = []

pencere = Tk()
pencere.geometry("200x50+600+460")

etiket = Label(fg = "white", bg = '#61380B', font = "Helvetica 12 bold")
etiket.pack()

for i in range(6):
    while len(liste) != 6:
        a = random.randint(1, 100)
        if a not in liste:
            liste.append(a)
            
etiket["text"] = liste

mainloop()
```

```python
from tkinter import *
import random

liste = []

class Uygulama():
    def __init__(self):
        self.araclar()
        self.kodlar()
        
    def kodlar(self):
        for i in range(6):
            while len(liste) != 6:
                a = random.randint(1, 100)
                if a not in liste:
                    liste.append(a)
        self.etiket["text"] = liste
        
    def araclar(self):
        self.etiket = Label(fg = "white", bg = "#61380B", font = 'Helvetica 12 bold')
        self.etiket.pack()
        
pencere = Tk()
pencere.geometry("200x50+600+460")
uyg = Uygulama()
mainloop()
```
